```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:",path)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

from sklearn import datasets
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
import pandas as pd
import os

# Assuming the CSV file is named 'creditcard.csv'
# and is located within the downloaded directory
file_path = os.path.join(path, 'creditcard.csv')

# Read the CSV file using the updated file path
df = pd.read_csv(file_path)
```

```python
X=df.drop(columns='Class',axis=1)
y=df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
model = LogisticRegression(max_iter= 500) # parameter called max_iter = 100 as default  (max_iteris the number of iterations.)
model.fit(X_train,y_train)
```

```python
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

print(accuracy_score(y_train, train_predictions))
print(accuracy_score(y_test, test_predictions))
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create an SVM model with a linear kernel
model = SVC(kernel='linear')
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Print classification report
print(classification_report(y_test, y_pred))
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
